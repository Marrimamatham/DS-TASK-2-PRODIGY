```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Load the dataset
task=pd.read_csv("Task 2.csv")
#Display the first few rows of the dataset
print("First few rows of the dataset:")
print(task.head())
```

```python
#Display shape
task.shape
```

```python
#checking for null values
task.isnull().sum()
```

```python
#Description of dataset
task.describe(include="all")
```

```python
task.groupby('Pclass')
```

```python
male_ind=len(task[task['Sex'] == 'male'])
print("No of males in Titanic:",male_ind)
```

```python
female_ind=len(task[task['Sex'] == 'female'])
print('No of females in Titanic:',female_ind)
```

## plotting 

```python
#plotting
figure = plt.figure()
axes = figure.add_axes([0,0,2,2])
gender = ['Male','Female']
index = [580,320]
axes.bar(gender,index,color='lightblue',edgecolor='black')
plt.xlabel("Gender")
plt.ylabel("No of people onboarding ship")
plt.show()
```

```python
#Distribution of age
plt.figure(figsize=(7,5))
plt.hist(task['Age'],bins=30,color='pink',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
alive=len(task[task['Pclass'] == 1])
dead = len(task[task['Pclass'] == 0])
task.groupby('Sex')[['Pclass']] . mean()
        
```

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])
status = ['Pclass','Dead']
ind = [alive,dead]
ax.bar(status,ind,color='red')
plt.xlabel("Status")
plt.show()
```

```python
plt.figure(1)
age=task.loc[task.Pclass==1,'Age']
plt.title('The histogram of the age groups of the people that had pclass')
plt.hist(age,np.arange(0,100,10),color='cyan')
plt.xticks(np.arange(0,100,10))
plt.figure(2)
age = task.loc[task.Pclass ==0,]
plt.title('The histogram of the age groups of the peoplethat coudn\'t pclass')
plt.hist(age,np.arange(0,100,10))
plt.xticks(np.arange(0,100,10))
```
